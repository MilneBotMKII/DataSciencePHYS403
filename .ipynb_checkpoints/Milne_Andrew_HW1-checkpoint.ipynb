```python
#Of people who get lung cancer, smokers are 25x more likely compared to non-smokers to get lung cancer
#If we are given a person who has lung cancer, what is the probability that they are a smoker?
#Probably need Bayes Theorum for this
#P(Smoker|Lung Cancer) = P(Lung Cancer| Smoker)P(Smoker)/P(Lung Cancer)

#P(Lung Cancer|Smoker) = 25P(Lung Cancer|NOT(Smoker)) is what I believe we are given.

#P(Lung Cancer) = P(Lung Cancer|Smoker)P(Smoker) + P(Lung Cancer|NOT(Smoker))P(NOT(Smoker)) 
#This is sorta sum rule? IDK what specifically it is
#But the probability in total must be the same as the sum of the conditional probabilities across all conditions

#Let's start subbing in. 
#P(Smoker|Lung Cancer) = P(Lung Cancer| Smoker)P(Smoker)/P(Lung Cancer) sub in P(Lung Cancer)
#P(Smoker|Lung Cancer) = P(Lung Cancer| Smoker)P(Smoker)/(P(Lung Cancer|Smoker)P(Smoker) + P(Lung Cancer|NOT(Smoker))P(NOT(Smoker))) sub in cancer rates
#P(Smoker|Lung Cancer) = P(Lung Cancer| Smoker)P(Smoker)/(P(Lung Cancer|Smoker)P(Smoker) + P(Lung Cancer|Smoker)P(NOT(Smoker))/25) P(Lung Cancer|Smoker) cancells
#P(Smoker|Lung Cancer) = P(Smoker)/(P(Smoker) + P(NOT(Smoker))/25) #Sub in P(NOT(Smoker)) = 1-P(Smoker)
#P(Smoker|Lung Cancer) = P(Smoker)/(P(Smoker) + 1-P(Smoker)/25)
#P(Smoker|Lung Cancer) = s/(s+((1-s)/25))

#I see 11.6% of adults in 2022 smoke (American Lung Association https://www.lung.org/research/trends-in-lung-disease/tobacco-trends-brief/overall-smoking-trends)

s=0.116
P = s/(s+((1-s)/25))
print('The probability of someone with lung cancer being a smoker is ' + str(100*P) + '%')
#Playing around with s, it's pretty interesting how small deviations affect the result
#But s=1 and s=0 give the correct answers and if you are 25x more likely to get cancer then effectively you have 25x11.6%=2.9 compared to 0.884 cases. 
#And doing the math on those unnormalized uhhh plausibilities we get the same numbers
```

```python
#Part C.)
#Let's simulate a bunch of random kid setups.
import numpy as np
from numpy import random as random
import matplotlib.pyplot as plt

ChanceForBlue=0.25 #1/4 chance
MaxFamilies = 100000 #Set a limit on the loop
FamilySize=3
Data=[]
temp=''
for i in range(MaxFamilies):
    for j in range(FamilySize):
        if (random.random() <= ChanceForBlue):
            temp+=('1') #I'm going to try doing this in a binary representation kinda just to figure it out
        else:
            temp+=('0')
    Data.append(int(temp, base=2))
    temp=''
        
#We now have a bunch of simulated families, we now can make a plot based on the value. There are 8 possible families
#I'm sure there is a better bitwise way to do this... like instead of Data being a list of lists it can be a list of binary numbers
#But idk how off the top of my head
fig, ax = plt.subplots()
counts, bins, bars = plt.hist(Data, bins = np.linspace(-1, 8, 10)+0.5)
plt.title('Simulated Families')
plt.ylabel('# Of Occourances')
plt.xlabel('Family Layout(Binary Representation)')
plt.xlim(-0.5, 7.5)
ax.set_xticklabels(['fix', '000', '001', '010', '011', '100', '101', '110', '111'])

#OK, now we can extract the raw numbers from our sample and compute probabilities. For instance with Part A.)
#We extract the data not in the 000 bin, which is all the families with at least 1 blue kid.
#Thankfully the hist function already has the counts. Little hardcodey but whatever

NumFamilesNoBlue = counts[0]
NumFamiliesAtLeastOneBlue = np.sum(counts[1:])
NumFamilesOneBlue = counts[1]+counts[2]+counts[4]
NumFamilesTwoBlue = counts[3]+counts[5]+counts[6]
NumFamiliesThreeBlue = counts[7]
NumFamiliesYoungestBlue = counts[1]+counts[3]+counts[5]+counts[7]
NumFamiliesOnlyYoungestBlue = counts[1]

#Part A.)
#We have the number of families with at least one blue, as well as the number of families with exactly 2 blue
#Just divide that out
print('The probability of a family having at least 2 blue eyed kids, given that we know one kid has blue eyes, is ' \
      + str((NumFamilesTwoBlue+NumFamiliesThreeBlue)/NumFamiliesAtLeastOneBlue * 100)+'%')

#Part B.)
#We know how many cases have the youngest child being blue
print('The probability of a family having at least 2 blue eyed kids, given that we know the youngest kid has blue eyes, is ' \
      + str((NumFamiliesYoungestBlue-NumFamiliesOnlyYoungestBlue)/NumFamiliesYoungestBlue * 100)+'%')
```
